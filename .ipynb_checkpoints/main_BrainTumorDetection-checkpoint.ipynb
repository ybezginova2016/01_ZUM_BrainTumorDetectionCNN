````python
# labeling the classes
classes = {'no' : 0, 'yes' : 1}
classes
````

````python
X=[]
y=[]

for cls in classes:
    pth = '/Users/yuliabezginova/PycharmProjects/00_files-for_NLP/Brain MRI Images for Brain Tumor Detection/brain_tumor_dataset/'+cls
    for j in os.listdir(pth):
        img = cv2.imread(pth+'/'+j, 0)
        img = cv2.resize(img, (200, 200))
        X.append(img)
        y.append(classes[cls])
````

````python
X = np.array(X)
y = np.array(y)
````
